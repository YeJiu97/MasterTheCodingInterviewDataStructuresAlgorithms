## Resume

这里有一个不错的网站：https://www.resumemaker.online/

这个在线的网络能够被用来制作简历。

点击resume之后会出现这样的一个网站：



这是一些关于简历的tips：
1. 一页A4纸，并且使用关键字
2. 展示相关的技能
3. 个性化/个人化
4. 提供网络链接（比如：Github链接，个人博客链接等）

更多的建议：



如果没有足够的工作经验，可以展示这些：
1. Github
2. Website
3. 1～2 大型项目
4. Bog

这是一份简单的简历：



发送简历：
1. 一次给多家公司发送简历
2. 对于几家有意愿的公司，进行更加细致的调查和针对性的优化简历

## LinkedIn

相对于投简历，也可以使用LinkedIn来展示自己：https://cn.linkedin.com/

## Portfolio

这部分可以用来展示自己的作品，可以自己创建一个网站，也可以使用在线网站展示自己的作品。

需要注意的是：
1. 不需要在这里花费太多的时间
2. 作品的组合需要具有多样性

## E-mail

不要使用学校的邮件，也不要使用QQ之类的邮件。

推荐使用outlook邮箱，并且邮件名称不能够是一连串的数字或者无意义的字符串，而需要是能够表示自己身份和容易记忆的名称。
