## Big O and Scalability

我们可以记录一下Python程序的运行时间：

```python
import datetime

start_time = datetime.datetime.now()

for _ in range(10000):
    for _ in range(10000):
        pass

end_time = datetime.datetime.now()

print(end_time - start_time)
```

我们可以再运行一次上述代码：

```python
import datetime

start_time = datetime.datetime.now()

for _ in range(10000):
    for _ in range(10000):
        pass

end_time = datetime.datetime.now()

print(end_time - start_time)
```

可以发现这段代码的运行速度非常的快，只需要1.28秒左右，但是我们也可以发现统一段代码两次的运行时间是不一样的。

这会给我们带来一个问题：同样的代码在相同的环境之下运行的时间存在着差异，在不同的环境之下更是会存在着差异，那么我们如何去评估一段代码的运行呢？

方法其实很简单：分析代码的运行次数。

我们会使用Big O来进行表示，O()。

这是几种常见的Big O的可视化结果：



## O(n)

我们可以看一下这样的一段代码：

```python
n = 10

for i in range(n):
    print(i+1, "次循环")
```

可以发现循环的次数为10。

接着我们将n的值进行修改，为原先的2倍，也就是20：

```python
n = 20

for i in range(n):
    print(i+1, "次循环")
```

可以发现循环的次数从10次变成了20次。

对于这一段代码而言，n的值发生了什么样的改变，循环的次数也就发生了什么样的改变。

如果n的值增加5,那么循环的次数也会增加5；如果n的值为原先的2倍，把么循环的次数也会增加2倍。



## O(1)

我们可以将上面的代码进行修改：

```python
n = 10

for i in range(5):
    print(n)
```

我们来试着修改一下n的值：

```python
n = 20

for i in range(5):
    print(n)
```

可以发现，n的值的改变并不会改变代码的运行次数。

当n为10的时候，代码运行了5次；当m为20的时候，代码还是运行了5次。无论n的值如何发生改变，代码的运行次数都不会发生任何的改变。



## O(n^2)

我们可以将前面的代码修改一下：

```python
n = 5
cout = 1

for i in range(n):
    for j in range(n):
        print(cout, "次循环")
        cout += 1
```

我们接着修改一下n的值：

```python
n = 10
cout = 1

for i in range(n):
    for j in range(n):
        print(cout, "次循环")
        cout += 1
```

可以发现n的值增加了2倍，代码运行的次数也就增加了2^2倍。

这并不难理解，因为这只不过是一个循环嵌套了另一个循环而以。

2. 移除常数
3. different terms of inputs
